```python
import pandas as pd
```

```python
custos_df = pd.read_csv("resultados/resultados_custo.csv")
custos_df.head()
```

```python
print("Custo médio de todas as instancias:")
aux_df = custos_df.iloc[:, 3:]
n, _ = aux_df.shape
for i in range(0, n):
    # desconsiderando os dois ultimos valores
    print(f"GRAPS: {custos_df.iloc[i, 1]:>3}, ILS: {custos_df.iloc[i, 2]:>3}, {aux_df.iloc[i, :-2].mean()}")
```
